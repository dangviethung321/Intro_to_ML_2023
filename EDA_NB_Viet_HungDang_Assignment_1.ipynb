```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data= pd.read_csv('./iris.csv')

data.info()
data.describe()
```

```python

import matplotlib.pyplot as plt
plt.figure()



plt.scatter(data['sepal_length'].tolist(),data['sepal_width'].tolist(),c = 'red')


plt.figure()
plt.plot(data['sepal_length'].tolist(),data['sepal_width'].tolist(),'bo')

```

```python
a = data['sepal_length'].tolist()
b = data['sepal_width'].tolist()
c = data['petal_length'].tolist()
d = data['petal_width'].tolist()

X = np.vstack((a,b))
X = np.vstack((X,c))
X = np.vstack((X,d))


correlation_matrix = np.corrcoef(X)
print('correlation_matrix \n',correlation_matrix)
```

```python
plt.figure()
plt.hist(a,bins=15)
plt.figure()
plt.hist(b,bins=15)
plt.figure()
plt.hist(c,bins=15)
plt.figure()
plt.hist(d,bins=15)
```
